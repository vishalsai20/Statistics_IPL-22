Let us remove the unwanted columns from the dataframe df and create a new dataframe

```python
df1 = df.drop(['MatchID','InningsNo','MatchPlayingOrder','BowlerName','OutDesc','StrikeRate','WicketNo','AgainstFast','AgainstSpin'],axis=1)
#The column StrikeRate is removed because it is of type 'Object'
df1
```

# Now let us Look at Team Performances

```python
agg_fun = {'Runs':'sum','Balls':'sum','DotBalls':'sum','Ones':'sum','Twos':'sum','Threes':'sum','Fours':'sum','Sixes':'sum'}

df2 = df1.groupby('Team').agg(agg_fun)
```

```python
df2
```

In this dataset number of matches played by each team is not given exclusively. So I am inserting a new row into dataframe df2 which shows number of matches played byeach team in IPL 2022

```python
df2.insert(0,'Matches Played',[14,14,16,14,15,14,14,16,17,14],True)
```

```python
df2
```

Let us plot graphs related to Team Performances

```python
plt.figure()
bar = plt.bar(list(df2.index.values),df2['Runs'])
plt.title('Runs v/s Teams')
plt.bar_label(bar,labels=df2['Runs'],label_type='edge')
plt.xlabel('Teams')
plt.ylabel('Runs scored')
```

```python
plt.figure()
plt.bar(list(df2.index.values),df2['Balls'])
plt.title('Balls Played v/s Teams')
plt.xlabel('Teams')
plt.ylabel('Balls Played')
```

```python
plt.figure()
plt.bar(list(df2.index.values),df2['DotBalls'])
plt.title('Dot Balls v/s Teams')
plt.xlabel('Teams')
plt.ylabel('Dotballs Played')
```

```python
plt.figure()
bar = plt.bar(list(df2.index.values),df2['Sixes'])
plt.title('Sixes v/s Teams')
plt.bar_label(bar,labels=df2['Sixes'],label_type='center')
plt.xlabel('Teams')
plt.ylabel('Sixes')
```

```python
plt.figure()
bar = plt.bar(list(df2.index.values),df2['Fours'])
plt.bar_label(bar,labels=df2['Fours'],label_type='center')
plt.title('Fours v/s Teams')
plt.xlabel('Teams')
plt.ylabel('Fours')
```

Here we had observed that team Rajasthan Royals has scored most number of runs (2807) in IPL 2022

Rajasthan Royals has also hit most number of Sixes (137)

Gurajat Titans has hit most number of Fours (250)

Now let us see Average Runs scored by each team in this edition.

```python
#Average runs scored in each match

avg_scr = []

for i in range(0,10):
    avg_scr.append(int(df2.iloc[i]['Runs']/df2.iloc[i]['Matches Played']))
    
```

Average Runs scored by each team 

```python
for i in range(0,10):
    print('Average Runs Scored by',list(df2.index.values)[i],'in each match :',avg_scr[i])
```

```python
plt.figure()
bar = plt.bar(list(df2.index.values),avg_scr)
plt.title('Average Runs Scored by each Team')
plt.bar_label(bar,labels=avg_scr,label_type='edge')
plt.xlabel('Teams')
plt.ylabel('Avg_Runs')
```

# Now let's look at individual performances

```python
agg_fun1 = {'Runs':'sum','Balls':'sum','DotBalls':'sum','Ones':'sum','Twos':'sum','Threes':'sum','Fours':'sum','Sixes':'sum'}

df3 = df1.groupby('PlayerName').agg(agg_fun1)
```

```python
df3
```

```python
df3.sort_values(by=['Runs'],ascending=False,inplace=True)
```

Using the above code to get the Players Name acoording to the Highest Runs scored

```python
df3
```

Top Ten Run Getters in IPL 22

```python
df5 = df3.iloc[0:10]
```

```python
df5
```

Plots with respect to Top Run Getters in IPL 2022

```python
plt.figure(figsize=(15,6))
bar = plt.bar(list(df5.index.values),df5['Runs'])
plt.title('Highest Run Getters')
plt.bar_label(bar,labels=df5['Runs'],label_type='edge')
plt.xlabel('Player Name')
plt.ylabel('Runs')
```

```python
plt.figure(figsize=(15,6))
bar = plt.bar(list(df5.index.values),df5['Fours'])
plt.title('Number of Fours')
plt.bar_label(bar,labels=df5['Fours'],label_type='edge')
plt.xlabel('Player Name')
plt.ylabel('Fours')
```

```python
plt.figure(figsize=(15,6))
bar = plt.bar(list(df5.index.values),df5['Sixes'])
plt.title('Number of sixes')
plt.bar_label(bar,labels=df5['Sixes'],label_type='edge')
plt.xlabel('Player Name')
plt.ylabel('Sixes')
```

Strike Rate of Top 10 Run Getters

```python
round(df5['Runs']/df5['Balls']*100,2)
```

Here it is clear that Jos Buttler has dominated all the batters in IPL 2022 by hitting most number of runs (863) which includes 45 sixes and 83 fours at a strike rate of 149.05.
